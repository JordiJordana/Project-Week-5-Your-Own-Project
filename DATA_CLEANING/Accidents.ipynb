```python
import pandas as pd
import numpy as np
```

## Read the files

```python
persones_17 = pd.read_excel('../2017_accidents_persones_gu_bcn_.xlsx')
persones_18 = pd.read_excel('../2018_accidents_persones_gu_bcn_.xlsx')
persones_18.head()
```

```python
persones_18.info()
```

## Append one year to another

```python
persones = pd.concat([persones_17, persones_18], ignore_index=True)
```

```python
persones.info()
```

## We change the name of the DataFrame, drop the unnecessary columns and change some of the names.

```python
crashes = persones
```

```python
cols=[1,3,4,5,6,7,9,22,24,25,26,27]
crashes.drop(crashes.columns[cols], axis=1, inplace=True)
```

```python
crashes.rename(columns={"Descripcio_dia_setmana": "Dia_setmana", "Descripcio_tipus_dia": "Tipus_dia", "Mes_ any": "Mes", "Descripcio_torn": "Torn", "Hora_dia": "Hora", "Descripcio_causa_vianant": "Causa", "Desc_Tipus_vehicle_implicat":"Vehicle_implicat","Descripcio_sexe":"Sexe", "Descripcio_tipus_persona":"Persona_afectada", "Descripcio_victimitzacio": "Victima"}, inplace=True)
crashes.head()
```

### In order to be able to use the 'Edat' column as a calculation, we turn the column into an integer and force the non convertible values into Nan.

```python
crashes['Edat'] = pd.to_numeric(crashes.Edat, errors='coerce')
crashes.Edat.unique()
```

```python
crashes.Edat.dtype
```

```python
crashes.Any = crashes.Any.apply(str)
```

```python
crashes["Dia_mes"] = crashes["Dia_mes"].map("{:02}".format)
crashes["Mes"] = crashes["Mes"].map("{:02}".format)
```

```python
crashes['Data'] = crashes.apply(lambda row: row.Any + '-' + row.Mes + '-' + row.Dia_mes, axis = 1)
crashes.head()
```

```python
crashes.Data = crashes.Data.astype('datetime64[ns]')
```

```python
crashes.to_csv('accidents.csv')
```
